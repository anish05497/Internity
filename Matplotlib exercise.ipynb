```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
Languages=['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
Popularity=[22.2, 17.6, 8.8, 8, 7.7, 6.7]
```

```python
xpos = np.arange(len(Languages))
xpos
```

```python
plt.bar(xpos,Popularity, label="Popularity")

plt.xticks(xpos,Languages)
plt.ylabel("Popularity")
plt.title('CS')
plt.legend()
```

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
men_means=[22, 30, 35, 35, 26]
women_means=[25, 32, 30, 35, 29]
menStd = [2, 3, 4, 1, 2]
womenStd = [3, 5, 2, 3, 3]
```

```python
ind = np.arange(len(men_means))
ind
```

```python
width=0.4
```

```python
plt.bar(ind-0.2,men_means, yerr=menStd, width=0.4, label="Men")
plt.bar(ind+0.2,women_means,yerr=womenStd, width=0.4,label="Women")

plt.xticks(ind,('G1', 'G2', 'G3', 'G4', 'G5'))
plt.yticks(np.arange(0, 81, 10))
plt.ylabel("Score")
plt.title('Scores by group and gender')
plt.legend()
```

```python
plt.bar(ind,men_means, yerr=menStd, width=0.4, label="Men")
plt.bar(ind,women_means,yerr=womenStd,bottom=men_means, width=0.4,label="Women")

plt.xticks(ind,('G1', 'G2', 'G3', 'G4', 'G5'))
plt.yticks(np.arange(0, 81, 10))
plt.ylabel("Score")
plt.title('Scores by group and gender')
plt.legend()
```
